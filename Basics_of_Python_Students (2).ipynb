- Lists are created using square brackets

<b>Some Python List characteristics: </b>
- List items are ordered, changeable, and allow duplicate values.
- List items are indexed, the first item has index [0], the second item has index [1] etc.
- Items  in a list have a defined order, and that order will not change. If you add new items to a list, the new items will be placed at the end of the list.
- The list is changeable, meaning that we can change, add, and remove items in a list after it has been created.
- Since lists are indexed and can have items with the same value

## Changing Tuples Values

Once a tuple is created, you cannot change its values. Tuples are unchangeable, or immutable as it also is called.
But there is a workaround. You can convert the tuple into a list, change the list, and convert the list back into a tuple.

````python
tuple_1 = ("apple", "banana", "cherry")  # taking a tuple
var_1 = list(tuple_1)                    # changing a tuple to a list 
var_1[1] = "kiwi"                        # changing a value in that list we created in above step
tuple_1 = tuple(var_1)                   # again changing that tuple to list
print(tuple_1)
````

````python
# we can add 2 tuples but not a single value to a tuple 
tuple_2 = ("apple", "banana", "cherry")
tuple_3 = ("Tomato",)
tuple_2 += tuple_3
print(tuple_2)
````

We can only use del() function on tuple. If we want to use any other function we have to change that to a list

## Unpacking Tuples

 When we create a tuple, we normally assign values to it. This is called "packing" a tuple, But, in Python, we are also allowed to extract the values back into variables. This is called "unpacking".

````python
countries = ("India","America","Russia")
(Safron,Blue,White) = countries # these values becomes a variable for the values of above values
print(Safron)
print(Blue)
print(White)
````

# <span style='color:blue'> Python Loops </span>

Python has two primitive loop commands:

- While loops

- For loops

## While Loop

With the while loop we can execute a set of statements as long as a condition is true.

````python
# e.g. we are printing hello_world as long as i is less than 10
i = 1
while i < 10:
    print("hello_world")
    i += 1
````

- Remember to increment the number, or else the loop will continue forever.

With the break statement we can stop the loop even if the while condition is true:

````python
# e.g. 
number = 10
while number < 15:
    print(i)
    if i == 12:
        break        # using break statement 
    i += 1
````

With the continue statement we can stop the current iteration, and continue with the next:

````python
num = 0
while num < 5:
    num += 1
    if num == 3:
        continue
    print(num)
````

With the else statement we can run a block of code once when the condition no longer is true:

````python
num1 = 1
while num1 < 6:
    print(num1)
    num1 += 1
else:
    print("num1 is no longer less than 6")
````

## For Loop

A for loop is used for iterating over a sequence. With the for loop we can execute a set of statements, once for each item in a list, tuple, set etc.

````python
colors = ["black","blue","green","red"]
for i in colors:     # taking i as reference of the values in the list colors
    print(i)
````

- Even strings are iterable objects, they contain a sequence of characters

````python
for i in "college":
    print(i)
````

- The range() function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and stops before a specified number.

````python
# e.g. 
for i in range(0,10): # we can also give 3rd parameter to skip the number of items in the range
    print("world")
````

````python
for i in range(0,11):
  print("abhi")
````

# <span style='color:blue'> Python Functions </span>

- A function is a block of code which only runs when it is called.
- You can pass data, known as parameters, into a function.
- A function can return data as a result.

In python function is defined using the def keyword:

````python
def func():      # defining a function 
    print("Hello world!")
# calling that function
func()
````

## Arguments in a function:

- Information can be passed into functions as arguments.
- Arguments are specified after the function name, inside the parentheses. We can add as many arguments as we want, just separate them with a comma.
- The following example has a function with one argument 'name'. When the function is called, we pass along a first name, which is used inside the function to print the full name

````python
def function_1(name):
    print(name + " world") 
function_1("Hello")
````

By default, a function must be called with the correct number of arguments. Meaning that if our function expects 2 arguments, we have to call the function with 2 arguments, not more, and not less.

If we do not know how many arguments that will be passed into our function, add a * before the parameter name in the function definition. This way the function will receive a tuple of arguments, and can access the items accordingly

````python
def function_2(*word):
    print("I love " + word[2])

function_2("c++", "java", "python")
````

## Passing a List as an Argument
We can send any data types of argument to a function (string, number, list, dictionary etc.), and it will be treated as the same data type inside the function.

````python
def function(food):
    for x in food:
        print(x)

fruits = ["apple", "banana", "cherry"]

function(fruits)
````

- If we want a function to return a value we can just write return in that function

````python
# e.g.
def function(x):
     return 10 * x

print(function(2))
print(function(5))
print(function(10))
````

## The pass Statement
Function definitions cannot be empty, but if we for some reason have a function definition with no content, put in the pass statement to avoid getting an error.

````python
def function():
    pass
````
