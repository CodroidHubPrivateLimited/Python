```python
import geopandas as gpd              #importing library 
import matplotlib.pyplot as plt
```

```python
gpd.datasets.available              #default datasets available in library


#kaggle can be used to import the dataset , but change the path accordingly
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))              #reading the file 
```

```python
world.head()
```

```python
world.plot()                    #plot is used to draw a graph
```

Here are some common columns in the Natural Earth low-resolution dataset:

1) pop_est: Population estimate
2) continent: Continent name
3) name: Country name
4) iso_a3: ISO Alpha-3 country code
5) gdp_md_est: GDP estimate in millions of dollars
6) geometry: Geometry (polygons representing the countries)

```python
f , ax = plt.subplots(1,1,figsize=(10,10))

world.plot(ax=ax,
           column = 'pop_est',
           cmap = 'RdYlGn_r')
```

```python
# Plot by country name (useful for checking country shapes)
fig, ax = plt.subplots(1, 1)
world.plot(ax=ax, column='pop_est', cmap='tab20', legend=False)
plt.show()
```

```python
from matplotlib.colors import LogNorm

norm = LogNorm(vmin = world.pop_est.min(),
               vmax = world.pop_est.max())

f , ax = plt.subplots(1,1,figsize=(10,5))

world.plot(ax=ax,
           column = 'pop_est',
           cmap = 'RdYlGn_r',
          edgecolor = 'w',
          linewidth = 1,
          legend = True,
          norm = norm)

ax.axis('off')                                                      #axis can be turned off

ax.set_title('Population per Country', fontsize = 14)               #title to describe what data the graph shows 
```
